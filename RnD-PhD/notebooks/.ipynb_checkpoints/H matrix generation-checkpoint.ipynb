```python
f = open(f'storage/H_matrix.py', 'w')

# cf {i_c} {i_f} {jj} {ii} {send_len} {take_len}

# >>>>>>>>>>>> Элемент матрицы H <<<<<<<<<<<<
with open(f'storage/docking_H_element_matrix.py', 'r') as s:
    txt = s.read()
txt = txt.replace('<to_replace>', addition_to_file())
f.write(txt)

GAIN_MODES = o.v.GAIN_MODES.copy()
GAIN_MODES.remove('3 antennas')
GAIN_MODES.remove('ellipsoid')
# >>>>>>>>>>>>

def H_element_sample(o, angles_navigation, gm_1, gm_2, multy_antenna_take, multy_antenna_send, relation):
    txt = f"    if (angles_navigation=={angles_navigation}) and (gm_1=='{gm_1}') and (gm_2=='{gm_2}') "
    txt += f"and (multy_antenna_send=={multy_antenna_send}) and (multy_antenna_take=={multy_antenna_take}) " 
    txt += f"and (relation=='{relation}'):\n"
    txt += f"        return vec_type([\n"
    
    G1 = kf.get_gain(v=o.v, obj=o.c, r=S1 @ (r2-r1), if_send=True, multake=multy_antenna_take, mulsend=multy_antenna_send, gm=gm_1)  # o.c/o.f роли не играет
    G2 = kf.get_gain(v=o.v, obj=o.f, r=S2 @ (r2-r1), if_take=True, multake=multy_antenna_take, mulsend=multy_antenna_send, gm=gm_2)
    
    if not multy_antenna_send and not multy_antenna_take:  
        G12 = [sum(G1)*sum(G2)]
        glen = 1
    if multy_antenna_send and multy_antenna_take: 
        G12 = [g1 * g2 for g1 in G1 for g2 in G2]
        glen = len(G12)

    if angles_navigation:
        z = [kf.sympy_norm(r1 - r2) / sqrt(gg) for gg in G12]  # Измерения
    else:
        z = [kf.sympy_norm(r1 - r2) for _ in G12]  # Измерения
           
    if angles_navigation:
        for row in range(glen):  # FOR по строкам (вниз-вверх)
            if relation == "cd":
                txt += f"{' '*23}["
                for ii in range(3):  # FOR по столбцам (влева-вправо)
                    txt += f"({my_simplify(z[row].diff(r2[ii]), angles_navigation)}), "
                for ii in range(3):  # FOR по столбцам (влева-вправо)
                    txt += f"({my_simplify(z[row].diff(q2[ii+1]), angles_navigation)}), "
            else:
                cnd1 = ""  # "*int(ff_sequence[i_y][0]==i_x)" # f"*int(i_n==i)*int(i_x==j)*int(ff_sequence[i_y][0]==i_x)"
                cnd2 = ""  # "*int(ff_sequence[i_y][1]==i_x)" # f"*int(i_n==j)*int(i_x==j)*int(ff_sequence[i_y][1]==i_x)"
                txt += f"{' '*23}["
                for ii in range(3):
                    txt += f"({my_simplify(z[row].diff(r1[ii]), angles_navigation)}){cnd1}+({my_simplify(z[row].diff(r2[ii]), angles_navigation)}){cnd2}, "
                for ii in range(3):
                    txt += f"({my_simplify(z[row].diff(q1[ii+1]), angles_navigation)}){cnd1}+({my_simplify(z[row].diff(q2[ii+1]), angles_navigation)}){cnd2}, "
            txt += f"] + 6*[0],\n"
    else:
        for row in range(glen):
            if relation == "cd":
                txt += f"{' '*23}["
                for ii in range(3):
                    txt += f"({my_simplify(z[row].diff(r2[ii]), angles_navigation)}), "
            else:
                cnd1 = ""  # "*int(ff_sequence[i_y][0]==i_x)" # f"*int(i_n==i)*int(i_x==j)*int(ff_sequence[i_y][0]==i_x)"
                cnd2 = ""  # "*int(ff_sequence[i_y][1]==i_x)" # f"*int(i_n==j)*int(i_x==j)*int(ff_sequence[i_y][1]==i_x)"
                txt += f"{' '*23}["
                for ii in range(3):
                    txt += f"({my_simplify(z[row].diff(r1[ii]), angles_navigation)}){cnd1}+({my_simplify(z[row].diff(r2[ii]), angles_navigation)}){cnd2}, "
            txt += f"] + 3*[0],\n"
    
    return txt + f"{' '*23}])\n\n"

n_vec = 6
tf = [True, False]
param_variation = [[angnav, gm_1, gm_2, mulsend, mulsend, rel] for angnav in tf for gm_1 in GAIN_MODES for gm_2 in GAIN_MODES for mulsend in tf for rel in ['cd', 'dd']]

anw = p_map(H_element_sample, 
            [o for _ in range(len(param_variation))], 
            [param_variation[i][0] for i in range(len(param_variation))],
            [param_variation[i][1] for i in range(len(param_variation))],
            [param_variation[i][2] for i in range(len(param_variation))],
            [param_variation[i][3] for i in range(len(param_variation))],
            [param_variation[i][4] for i in range(len(param_variation))],
            [param_variation[i][5] for i in range(len(param_variation))])
```
