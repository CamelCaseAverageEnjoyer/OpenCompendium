```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sympy import *
from __init__ import *

o = kf.init()
o.v.ORBIT_RADIUS = 6371e3 + 600e3
o.v.W_ORB = np.sqrt(o.v.MU / o.v.ORBIT_RADIUS ** 3)

def a_orb(w, x, y, z, vx, vy, vz):
    return kf.vec_type([3*w**2*x + 2*w*vy,
                        -2*w*vx,
                        -w**2*z])

# Скоропировано из "Решений уравнений"
def A(w, t):
    return kf.vec_type([[4 - 3*kf.cos(w*t), 0, 0],
                        [-6*w*t + 6*kf.sin(w*t), 1, 0],
                        [0, 0, kf.cos(w*t)]])

def B(w, t):
    return kf.vec_type([[kf.sin(w*t), -2*kf.cos(w*t) + 2, 0],
                        [2*kf.cos(w*t) - 2, 4*kf.sin(w*t) - 3*w*t, 0],
                        [0, 0, kf.sin(w*t)]]) / w

def C(w, t):
    return kf.vec_type([[3*w*kf.sin(w*t), 0, 0],
                        [-6*w + 6*w*kf.cos(w*t), 0, 0],
                        [0, 0, -w*kf.sin(w*t)]])

def D(w, t):
    return kf.vec_type([[kf.cos(w*t), 2*kf.sin(w*t), 0],
                        [-2*kf.sin(w*t), 4*kf.cos(w*t) - 3, 0],
                        [0, 0, kf.cos(w*t)]])


def get_C(w, x, y, z, vx, vy, vz):
    return [-6*x*w - 3*vy, 
            2*vx,
            6*x*w + 4*vy,
            y - 2*vx/w,
            vz,
            z*w]

def r_hkw(w, t, c1, c2, c3, c4, c5, c6):
    return kf.vec_type([
        (-4*c1 + 3*c2*kf.sin(t*w) - 3*c3*kf.cos(t*w)) / (6*w),
        (c2*kf.cos(t*w) + c3*kf.sin(t*w) + w*(c1*t + c4)) / w,
        (c5*kf.sin(t*w) + c6*kf.cos(t*w)) / w
    ])

def v_hkw(w, t, c1, c2, c3, c4, c5, c6):
    return kf.vec_type([
        c2*kf.cos(t*w)/2 + c3*kf.sin(t*w)/2,
        c1 - c2*kf.sin(t*w) + c3*kf.cos(t*w),
        c5*kf.cos(t*w) - c6*kf.sin(t*w)
    ])
```

```python
N_heatmap = 50
rate = 3
dx = 2
x_list = np.linspace(-dx, dx, N_heatmap)
y_list = np.linspace(-dx, dx, N_heatmap)

dr = np.array([0.1, 0.1, 0])
d = 0.5
ro1 = np.array([-1, 0, 0])
ro2 = np.array([1, 0, 0])
r1 = np.array([1, 1, 0])

def h(r1, r2, ro1, ro2):
    tmp = np.dot((ro2-ro1)/np.linalg.norm(ro2-ro1),(r1+r2)/2-ro1)
    r = (r2+r1)/2
    ro = (ro2+ro1)/2
    # if 0 < tmp and tmp < np.linalg.norm(ro2-ro1):
    if (0 < tmp and tmp < np.linalg.norm(ro2-ro1)):
        if (np.linalg.norm(r-ro) <= np.linalg.norm(r1-ro)) and (np.linalg.norm(r-ro) <= np.linalg.norm(r2-ro)):
            a = ro1 - r1
            b = r2 - r1
            c = ro2 - ro1
            # print(f"a: {a}, {b}, {c} | {np.cross(b, c)}")
            return np.dot(a,np.cross(b,c)) / np.linalg.norm(np.cross(b, c))
        else:
            return 10
    else:
        a = np.linalg.norm(r1 - ro1)
        b = np.linalg.norm(r1 - ro2)
        c = np.linalg.norm(r2 - ro1)
        d = np.linalg.norm(r2 - ro2)
        # print(f"b: ({a}, {b}, {c}, {d}) | {r1}, {r2}")
        return min(a, b, c, d)

def k(x_):
    return (x_ / (x_list[-1] - x_list[0]) + (x_list[-1] + x_list[0])/2 + 0.5) * N_heatmap
def k_(x_):
    return ((x_ / N_heatmap) - 0.5 - (x_list[-1] + x_list[0])/2) * (x_list[-1] - x_list[0])
    # return (x_ / (x_list[-1] - x_list[0]) + (x_list[-1] + x_list[0])/2 + 0.5) * N_heatmap
        
anw = [[h(r1=np.array([k_(ix),k_(iy),0]), 
          r2=np.array([k_(ix),k_(iy),0]) + dr,
          ro1=ro1, ro2=ro2) for ix in range(N_heatmap)] for iy in range(N_heatmap)]


fig, axes = plt.subplots(1, 1, figsize=(10, 8))
v_x_list = np.linspace(-dx, dx, N_heatmap)
v_z_list = np.linspace(-dx, dx, N_heatmap)
xlabels_0 = ['{:4.2f}'.format(x) for x in v_z_list]
ylabels_0 = ['{:4.2f}'.format(y) for y in v_x_list]
ax = sns.heatmap(anw, ax=axes, cmap="crest", xticklabels=xlabels_0, yticklabels=ylabels_0,  cbar_kws={'label': 'Distance h, m'})
ax.set_xticks(ax.get_xticks()[::rate])
ax.set_xticklabels(xlabels_0[::rate])
ax.set_yticks(ax.get_yticks()[::rate])
ax.set_yticklabels(ylabels_0[::rate])
ax.plot([k(ro1[0]), k(ro2[0])], [k(ro1[1]+d), k(ro2[1]+d)], c='c')
ax.plot([k(ro1[0]), k(ro2[0])], [k(ro1[1]-d), k(ro2[1]-d)], c='c')
ax.plot([k(ro1[0]), k(ro1[0])], [k(ro1[1]-d), k(ro1[1]+d)], c='c')
ax.plot([k(ro2[0]), k(ro2[0])], [k(ro2[1]-d), k(ro2[1]+d)], c='c')
ax.plot([k(ro1[0]), k(ro2[0])], [k(ro1[1]), k(ro2[1])], c='b')
ax.set_xlabel(f"x, m")
ax.set_ylabel(f"y, m")
plt.show()
    
```

##### <u>Управление</u>

```python
# Параметры
dt, T = 1., 5000
amax = 1e-4
w0 = o.v.W_ORB
t = np.linspace(0, T, int(T // dt))

# Целевая траектория
r0 = np.array([2, 3, 4])
v0 = np.array([0.1, 0.2, -0.1])
# dr0 = np.array([0.1, 0.1, 0.1])
dv0 = np.array([0.01, 0.01, 0.01])
v0[1] = -2*r0[0]*w0  # Поправка на нулевой дрейф
kf.my_print(f"Положение: {r0} | Скорость: {v0}", bold=True)

r1 = A(w0, T) @ r0 + B(w0, T) @ v0
v1 = C(w0, T) @ r0 + D(w0, T) @ v0

rt = np.zeros([int(T // dt), 3])
rn = np.zeros([int(T // dt), 3])
for i in range(int(T // dt)):
    rt[i] = A(w0, i*dt) @ r0 + B(w0, i*dt) @ v0
    rn[i] = A(w0, i*dt) @ (r0 + dr0) + B(w0, i*dt) @ (v0 + dv0)

def my_show_3d(axes):
    axes.scatter(r0[0], r0[1], r0[2], c='b', label='Start')
    axes.scatter(r1[0], r1[1], r1[2], c='r', label='Finish')
    axes.legend()
    axes.set_xlabel("x, m")
    axes.set_ylabel("y, m")
    axes.set_zlabel("z, m")
    plt.show()

axes = plt.figure(figsize=(6, 6)).add_subplot(projection='3d')
axes.plot(rt[:,0], rt[:,1], rt[:,2], label='Target trajectory')
axes.plot(rn[:,0], rn[:,1], rn[:,2], label='No-fuel trajectory')
my_show_3d(axes)
```

###### *Обычное сведение коэффициентов хку*

```python
k_r, k_v, k_i = 1e-4, 1e-3, 1e-15
xi = r1
vi = v1  
# vi = np.zeros(3)
ci = get_C(w0, *xi, *vi)

r_control = rn.copy()
v_control = v0 + dv0*1
integ = np.zeros(3)
u_list = np.zeros([int(T // dt), 3])

for i in range(1, int(T // dt)):
    a1 = a_orb(w0, *r_control[i-1], *v_control)
    
    r_target = r_hkw(w0, t[i]-T-dt, *ci)
    v_target = v_hkw(w0, t[i]-T-dt, *ci)
    integ += (r_target - r_control[i-1]) * dt
    a2 = k_i * integ + k_r * (r_target - r_control[i-1]) + k_v * (v_target - v_control)
    # a2 -= a1
    a2 = a2 if np.linalg.norm(a2) < amax else a2 * amax / np.linalg.norm(a2)
    u_list[i] = a2

    v_control += (a1 + a2)*dt
    # v_control += (a1)*dt  # Для отладки включить
    r_control[i] = r_control[i-1] + v_control*dt

axes = plt.figure(figsize=(6, 6)).add_subplot(projection='3d')
axes.plot(rt[:,0], rt[:,1], rt[:,2], label='Target trajectory')
axes.plot(rn[:,0], rn[:,1], rn[:,2], label='No-fuel trajectory')
axes.plot(r_control[:,0], r_control[:,1], r_control[:,2], label='Controlled trajectory')
my_show_3d(axes)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 4))

axs[1].plot(t, [np.linalg.norm(u_list[i, :]) for i in range(len(t))], 'k', label='u')
for i in range(3):
    axs[0].plot(t, r_control[:, i] - rt[:,i], label=['Δx', 'Δy', 'Δz'][i])
    axs[1].plot(t, u_list[:, i], label=['uˣ', 'uʸ', 'uᶻ'][i])
for i in range(2):
    axs[i].set_xlabel("time, s")
    axs[i].grid()
    axs[i].set_title(["coordinate error, m", "acceleration, m/s²"][i])
    axs[i].legend()

kf.my_print(f"Максимальное ускорение: {amax} м/с²", bold=True)
kf.my_print(f"Для 5 кг тяга двигателя: {(amax * 5)*1e3} мН\n")
plt.show()
```

```python
def v_gath(w: float, t: float, xi, r, T):
    tmp = np.zeros(3)
    B_1 = np.linalg.inv(B(w=w, t=t))
    B1 = np.linalg.inv(B(w=w, t=(T - t)))
    A_1 = A(w=w, t=t)
    A1 = A(w=w, t=(T - t))
    # print(f"1: {B_1 @ xi} | 2: {B_1 @ A1 @ r}")
    return B_1 @ xi - B_1 @ A1 @ r
    # return - B_1 @ A1 @ r

xi = r1

r_control = rn.copy()
v_control = v0 + dv0
u_list = np.zeros([int(T // dt), 3])

for i in range(1, int(T // dt)):
    a1 = a_orb(w0, *r_control[i-1], *v_control)
    
    v_target = v_gath(w=w0, t=t[i], xi=xi, T=T, r=r_control[i-1]) if i < int(T // dt) - 1 else np.zeros(3)  # Тут зарыт костыль
    a2 = (v_target - v_control) / dt
    a2 -= a1
    a2 = a2 if np.linalg.norm(a2) < amax else a2 * amax / np.linalg.norm(a2)
    u_list[i] = a2

    v_control += (a1 + a2)*dt
    # v_control = v_target  # Для отладки включить
    # v_control += (a1)*dt  # Для отладки включить
    r_control[i] = r_control[i-1] + v_control*dt

axes = plt.figure(figsize=(6, 6)).add_subplot(projection='3d')
axes.plot(rt[:,0], rt[:,1], rt[:,2], label='Target trajectory')
axes.plot(rn[:,0], rn[:,1], rn[:,2], label='No-fuel trajectory')
axes.plot(r_control[:,0], r_control[:,1], r_control[:,2], label='Controlled trajectory')
my_show_3d(axes)
```
