##### №5: Газгольдер

```python
class GasHolder:
    # Конструктор, принимающий один параметр - ёмкость газгольдера.
    # При попытке создать газгольдер отрицательного объёма нужно выбросить исключение (любое).
    # Резервуар создаётся пустой.
    # Температура созданного резервуара равна 273 К.
    def __init__(self, V):
        if V < 0:
            raise ValueError
        self.V = V
        self.T = 273
        self.R = 8.31
        self.gas = {}  # Молярная масса : масса

    # Впрыск порции газа массой m и молярной массой M.
    # При попытке передать отрицательные значения нужно выбросить исключение (любое).
    # Считать, что газ принимает текущую температуру газгольдера за пренебрежимо малое время.
    def inject(self, m, M):
        if m < 0 or M < 0:
            raise ValueError
        if M in self.gas.keys():
            self.gas[M] += m
        else:
            self.gas[M] = m

    # Подогреть газгольдер на dT градусов.
    # Считать, что нагрев возможен до любых значений температуры.
    def heat(self, dT):
        self.T += dT

    # Охладить газгольдер на dT градусов.
    # При попытке охладить ниже 0 К температура становится ровно 0 К.
    def cool(self, dT):
        self.T -= dT
        if self.T < 0:
            self.T = 0

    # Получить текущее давление в газгольдере.
    # Считать, что для газа верно уравнение состояния PV = (m/M)RT. 
    # Значение постоянной R принять 8.31 Дж/(моль*К).
    def get_pressure(self):
        tmp = 0
        for M in self.gas.keys():
            tmp += self.gas[M] / M
        return tmp * self.R * self.T / self.V
```

```python
h = GasHolder(1.0)
h.inject(29, 29)
print(f'Pressure after operation: {h.get_pressure()} Pa')
h.inject(29, 29)
print(f'Pressure after operation: {h.get_pressure()} Pa')
h.heat(273)
print(f'Pressure after operation: {h.get_pressure()} Pa')
h.cool(373)
print(f'Pressure after operation: {h.get_pressure()} Pa')
h.cool(373)
print(f'Pressure after operation: {h.get_pressure()} Pa')
```

##### №7: Антиспам

```python
input_1 = ["8 999 111 11 11, 8 999 111-11-22, 8 (999) 111 11 33",
           "8(999)111-11-11, 89991111122, 8 (999) 1111133",  
           "+7(999)111-11-11, +79991111122, +7 (999) 1111144"]  # Источник о спамерах
input_2 = "+7 (999) 222-22-11, +7 (999) 222-22-22, +7 (999) 111-11-22"  # Данные телефонной книжки
input_3 = "+7(999)111-11-22" # Номер входящего звонка
```

```python
input_1 = ["8 999 111 11 11, 8 999 111-11-22, 8 (999) 111 11 33",
           "8(999)111-11-11, 89991111122, 8 (999) 1111133",  
           "+7(999)111-11-11, +79991111122, +7 (999) 1111144"]  # Источник о спамерах
input_2 = "+7 (999) 222-22-11, +7 (999) 222-22-22, +7 (999) 111-11-22"  # Данные телефонной книжки
input_3 = "8 999 111-11-33" # Номер входящего звонка
```

```python
def clear_number(n):
    anw = n.replace(" ", "").replace("-", "").replace(")", "").replace("(", "")
    anw = anw[1:] if anw[0] == "8" else anw[2:]  # Убираем 8- или +7-
    return anw

spam_sources = [[clear_number(j) for j in input_1[i].split(",")] for i in range(3)]
home_sources = [clear_number(j) for j in input_2.split(",")]
recieved_num = clear_number(input_3)

anw = "NOT SPAM"

# Проверка на источники спамеров
counter = 0
for i in range(3):
    if recieved_num in spam_sources[i]:
        counter += 1
if counter > 1:
    anw = "SPAM"

# Проверка на телефонную книжку
if recieved_num in home_sources:
    anw = "NOT SPAM"

print(anw)
```

##### №8: KNN

```python
input_1 = "3"
input_2 = "5"
input_3 = ["1.92 1.6 0.56",
           "2.79 1.92 0.24",
           "2.36 1.89 0.15",
           "1.93 1.5 0.5",
           "1.15 1.6 0.25"]
input_4 = "0 1 1 0 0"
input_5 = "2.63 2.02 0.2"
```

```python
N = int(input_1)
M = int(input_2)

def get_dist(a, b):
    return (a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2 

a = []
params = list(map(float, input_5.split()))
for i in range(M):
    a.append(list(map(float, input_3[i].split())) + [int(input_4.split()[i])])
    a[i].append(get_dist(params, a[i]))
    
a = sorted(a, key=lambda x: x[4])

for i in a:
    print(f"животное: {i}")

anw = 0
for i in range(N):
    anw += a[i][3] / N

print(f"Ответ: {anw}, или {'elephant' if anw > 0.5 else 'hippopotamus'}")
```
