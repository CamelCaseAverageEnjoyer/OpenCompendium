##### Функции

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

table = pd.read_excel("files/Python-осень-2024.ods")
names = ["Гидра", "Максимальный палиндром", "Контрольная", "Крестики-нолики", "Газгольдер", "Ёжик в матане", "Антиспам", "KNN"]
colors = ['c', 'm', 'navy', 'g', 'orange', 'y', 'tan', 'gray']
group_name = 'Студенты  Б03-303'
task_amount = 8
```

```python
columns = {}
counter = 0
for task in range(1, task_amount+1):
    for attribute in ["N-attempt", "Time-start", "Time-finished", "Time-delta", "Done", "Style"]:
        counter += 1
        columns[table.columns[counter]] = f"{task} {attribute}"

table = table.rename(columns=columns)
table = table.drop([0, 1])

students = list(table[group_name])

indexes = {}
for i, s in zip(range(len(students)), students):
    indexes[i + 2] = s
table = table.rename(index=indexes)
```

```python
def time2sec(t):
    return t.hour * 60 + t.minute

def sec2time(t):
    pass

def show_time_complete(student: str):
    for task in range(1, 9):
        fig, ax = plt.subplots(figsize=(10, 3))
        print(f"\033[1mЗадача {task}: {names[task - 1]}\033[0m")
        students_now = []
        for i, s in zip(range(len(students)), students):
            if table[f"{task} Done"][i] == 1:
                students_now.append(s)
                c = "r" if student[1] in s and student[1] != "" else ("g" if student[0] in s and student[0]!="" else colors[task - 1])
                size = 50 if (student[0] in s and student[0] != "") or (student[1] in s and student[1] != "") else 20
                ax.scatter([time2sec(table[f"{task} Time-start"][i]), 
                            time2sec(table[f"{task} Time-finished"][i])], [s, s], lw=1, c=c, s=size)
                ax.plot([time2sec(table[f"{task} Time-start"][i]), 
                         time2sec(table[f"{task} Time-finished"][i])], [s, s], lw=1, c=c)
    
        for i, s in zip(range(len(students)), students):
            if s in students_now:
                ax.plot([time2sec(table[f"{task} Time-finished"][i]), 
                         time2sec(table[f"{task} Time-finished"][i])], 
                         [students_now[0], students_now[-1]], c=colors[task - 1])
    
        # plt.legend()
        plt.gca().invert_yaxis()
        plt.grid()
        plt.show()

def show_first_attempt(min_fast_received: int = 3, student: str = None):
    for i, s in zip(range(len(students)), students):
        attempt_list = []
        task_list = []
        for task in range(1, 9):
            if table[f"{task} Done"][i] == 1:
                attempt_list.append(table[f"{task} N-attempt"][i])
                task_list.append(task)
    
        color = None
        if attempt_list == [1 for _ in range(len(task_list))]:
            color = 'k' 
    
        if (attempt_list.count(1) > min_fast_received if student is None else student in s):  
            fig, ax = plt.subplots(figsize=(14, 2))
            
            tmp = [f'{task_list[j]}({attempt_list[j]})' for j in range(len(task_list))]
            print(f"\033[1mПопытки <{(s + '>:').ljust(30)}", end='')
            print(*tmp, end='')
            print(f"\033[0m")
    
            for j in range(len(task_list)):
                if attempt_list[j] == 1:
                    task = task_list[j]
                    print(f"Задачу {task} к тому моменту решили:")
                    flag = True
                    for ii, ss in zip(range(len(students)), students):
                        if table[f"{task} Done"][ii] == 1:
                            t1 = time2sec(table[f"{task} Time-finished"][ii]) 
                            t2 = time2sec(table[f"{task} Time-finished"][i])
                            if t1 < t2:
                                print(" "*10 + f"{ss} (на {t2 - t1} минут раньше)")
                                flag = False
                    if flag:
                        print(" "*10 + f"Никто!")
    
            for t, a in zip(task_list, attempt_list):
                ax.scatter(t, a, lw=1, c='c' if a > 1 else 'r', label=s)
            ax.plot(task_list, attempt_list, lw=1, c=color)
    
            # plt.rcParams["figure.figsize"] = (10,2)
            ax.set_xlabel("Задача")
            ax.set_ylabel("С какой попытки")
            plt.grid()
            # plt.legend()
            plt.show()

def show_same_done():
    tasks_done_8 = []
    for i, s in enumerate(students):
        tasks_done = []
        for task in range(1, 9):
            tasks_done.append(table[f"{task} Done"][s])
        tasks_done_8.append(tasks_done)
            
    for n in range(256):
        flag = True
        tasks_done = [int(i) for i in bin(n)[2:].rjust(8, "0")]
        for i, s in enumerate(students):
            if tasks_done == tasks_done_8[i]:
                if flag:
                    flag = False
                    print(f"Выполненные задачи \033[1m{tasks_done}\033[0m:")
                print(f"                   > {s}")

def show_solves(problem: int, student: str):
    path = f"files/test-1/2024-autumn-python-1-v2/problem_{problem}/"
    attempts = []
    for s in os.listdir(path):
        if student + "_" in s:
            attempts.append(s)
    attempts = sorted(attempts)
    
    s_name, _ = attempts[0].split("-")
    s_name = " ".join(s_name.split("_"))
    print(f"\033[1mСтудент: {s_name}\033[0m")
    
    times = []
    for a in attempts:
        _, t = a.split("time_")
        t, _ = t.split(".")
        times.append(":".join(t.split("_")))
    
    for i in range(len(times)):
        with open(path + attempts[i]) as f:
            print(f"\033[1mВремя решения: {times[i]}\033[0m")
            # print(f.read())
            for line in f:
                print(" "*5 + ">" + " "*2 + line.replace("\n", ""))

def compare(problem, student_1, student_2):
    if student_1 == student_2:
        show_solves(problem = problem, student = student_1)
    else:
        if student_1 not in [None, ""]:
            show_solves(problem = problem, student = student_1)
        if student_2 not in [None, ""]:
            show_solves(problem = problem, student = student_2)
```
